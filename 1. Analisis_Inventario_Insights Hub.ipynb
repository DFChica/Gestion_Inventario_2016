```python
#El método pd.to_datetime() de Pandas se usa para convertir la columna InvoiceDate del DataFrame purches a un formato de fecha y hora (datetime). 
#Este tipo de conversión es crucial para trabajar con datos temporales, ya que facilita su análisis, ordenación y manipulación.
purches['InvoiceDate'] = pd.to_datetime(purches['InvoiceDate'])
```

```python
# Este script realiza una operación de suma entre dos columnas del DataFrame purches y almacena el resultado en una nueva columna llamada Total Amount.
purches['Total Amount'] = purches['Dollars'] + purches['Freight']
```

```python
# purches.head(2) muestra las primeras 2 filas del DataFrame purches.
purches.head(2)
```

```python
# El método purches['Approval'].unique() devuelve los valores únicos presentes en la columna Approval del DataFrame purches.
purches['Approval'].unique()
```

```python
# El comando purches.drop(drop, axis = 1, inplace = True) elimina las columnas especificadas en la lista drop del DataFrame purches.
drop = ['VendorNumber','PONumber','PODate']
purches.drop(drop, axis = 1, inplace =True)
```

```python
# Este script realiza varias acciones en el DataFrame purches para encontrar el proveedor (vendor) asociado con la compra de mayor cantidad 
# y luego imprime el nombre del proveedor junto con la cantidad máxima comprada.
High_purchases_item =purches.loc[purches['Quantity'].idxmax(), 'VendorName']
Quty = purches['Quantity'].max()
print("High Prchuse Vendor Name:- ",High_purchases_item,"Quantity :-", Quty)
```

```python
# Este script realiza dos operaciones sobre el DataFrame purches y se enfoca en el proveedor DIAGEO NORTH AMERICA INC. La primera operación cuenta el número total de facturas (invoices) relacionadas con ese proveedor,
# y la segunda calcula el gasto total (spend) realizado con ese proveedor.
total_num_of_invoice = purches[purches['VendorName'] == 'DIAGEO NORTH AMERICA INC   '].shape[0]
print('Total Number Of Invoice:-',total_num_of_invoice)
Spend_Total_Amount =purches.loc[purches['VendorName'] == 'DIAGEO NORTH AMERICA INC   ','Total Amount'].sum()
print("Spend Total Amount:-",Spend_Total_Amount)
```

```python
# Este script realiza varias operaciones sobre el DataFrame purches para identificar el proveedor (vendor)
# asociado con la compra de menor cantidad y luego imprime el nombre del proveedor junto con la cantidad mínima comprada.
Low_purchases_item =purches.loc[purches['Quantity'].idxmin(), 'VendorName']
Quty2 = purches['Quantity'].min()
print("Low Prchuse Vendor Name:- ",Low_purchases_item,"Quantity :-", Quty2)
```

```python
# Este script realiza un cálculo sobre el DataFrame purches para calcular el gasto total realizado con un proveedor específico, 
# en este caso HOOD RIVER DISTILLERS, Inc., y luego imprime ese gasto total.
Spend_Total_Amount =purches.loc[purches['VendorName'] == 'HOOD RIVER DISTILLERS, Inc.','Total Amount'].sum()
print("Spend Total Amount:-",Spend_Total_Amount)
```

```python
# Este script elimina ciertas columnas del DataFrame llamado sales.
drop1 = ['InventoryId','VendorNo','Brand',]
sales.drop(drop1,axis=1,inplace=True)
```

```python
# Este script crea una nueva columna en el DataFrame sales llamada 'Total Amount'**.
# La columna **'Total Amount'** se calcula sumando los valores de dos columnas existentes: 'SalesDollars' y 'ExciseTax'`.
sales['Total Amount'] = sales['SalesDollars'] + sales['ExciseTax']
```

```python
# Este script está diseñado para encontrar el proveedor con la máxima cantidad de ventas en el DataFrame sales
# y mostrar el nombre del proveedor junto con la cantidad de ventas correspondiente.
High_sales_item =sales.loc[sales['SalesQuantity'].idxmax(), 'VendorName']
Quty3 = sales['SalesQuantity'].max()
print("High sales Vendor Name:- ",High_sales_item,"Quantity :-", Quty3)
```

```python
# Este script calcula y muestra dos métricas relacionadas con el proveedor 'E & J GALLO WINERY' en el DataFrame sales:
# Número total de facturas para ese proveedor.
# Monto total gastado con ese proveedor.
total_num_of_invoice3 = sales[sales['VendorName'] == 'E & J GALLO WINERY         '].shape[0]
print('Total Number Of Invoice:-',total_num_of_invoice3)
Spend_Total_Amount3 =sales.loc[sales['VendorName'] == 'E & J GALLO WINERY         ','Total Amount'].sum()
print("Spend Total Amount:-",Spend_Total_Amount3)
```

```python
# Este script se utiliza para identificar el proveedor con la menor cantidad de ventas en el DataFrame sales.
# Es útil en análisis de ventas para saber cuál proveedor ha tenido el peor desempeño en términos de cantidad vendida.
Low_sales_item =sales.loc[sales['SalesQuantity'].idxmin(), 'VendorName']
Quty4 = sales['SalesQuantity'].min()
print("Low sales Vendor Name:- ",Low_sales_item,"Quantity :-", Quty4)
```

```python
#Este script se utiliza para calcular y obtener el monto total gastado con un proveedor específico ('JIM BEAM BRANDS COMPANY' en este caso).
# Es útil en análisis financieros para evaluar el gasto total realizado con un proveedor determinado dentro de un conjunto de ventas.
Spend_Total_Amount4 =sales.loc[sales['VendorName'] == 'JIM BEAM BRANDS COMPANY    ','Total Amount'].sum()
print("Spend Total Amount:-",Spend_Total_Amount4)
```

```python
# Este gráfico te permitirá ver cómo se distribuyen los valores de Total Amount en el conjunto de datos,
# lo cual puede ser útil para identificar patrones, como distribuciones sesgadas o concentradas.

import seaborn as sns
import matplotlib.pyplot as plt

# Histograma de 'Total Amount'
plt.figure(figsize=(10, 6))
sns.histplot(sales['Total Amount'], bins=30, kde=True, color='blue')
plt.title('Distribución de Monto Total')
plt.xlabel('Monto Total')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.tight_layout()
plt.show()


```
